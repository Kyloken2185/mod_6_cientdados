```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

```python
df = pd.read_csv('dados.csv')
```

```python
df1 = df.copy()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Boxplot da idade da mãe:

sns.boxplot(data = df1, y ='IDADEMAE', palette = 'Set2')
plt.ylabel('Idade da mãe')

# Exibir o gráfico
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Boxplot da idade da mãe:

sns.boxplot(data = df1, y ='PESO', palette = 'Set2')
plt.ylabel('Peso do bebê')

# Exibir o gráfico
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
df1['IDADEMAE'] = df1['IDADEMAE'].fillna(0)
sns.displot( data = df1
            , alpha = .5
            , x = 'IDADEMAE'
            , bins = 5
            , element = 'step'
            , kde = True )
plt.xlabel('Idade da mãe')
plt.ylabel('Quantidade de mulheres')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
df1['PESO'] = df1['PESO'].fillna(0)
sns.displot( data = df1
            , alpha = .5
            , x = 'PESO'
            , bins = 5
            , element = 'step'
            , kde = True )
plt.xlabel('Peso da criança')
plt.ylabel('Quantidade de crianças')
plt.show()
```
